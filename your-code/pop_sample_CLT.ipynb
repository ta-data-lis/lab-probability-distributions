```python
import numpy as np
import matplotlib.pyplot as plt
```

# Sample Mean variation

```python
### randomly generate a population of ages of people
population=np.random.randint(0,95,10000)

np.mean(population)
```

```python
##small-ish sample size - large variation due to sampling

pop_sample=np.random.choice(population,10)#Picks 10 elements at random
np.mean(pop_sample)
```

```python
##larger sample size - large variation due to sampling

pop_sample=np.random.choice(population,100)#Picks 100 elements at random
np.mean(pop_sample)
```

```python
true_sigma=population.std(ddof=0)
true_sigma
```

```python
pop_sample=np.random.choice(population,10)
pop_sample.std(ddof=1)
```

```python
pop_sample=np.random.choice(population,100)
pop_sample.std(ddof=1)
```

```python
#plotting the popullation distribution
plt.hist(population)
plt.show()
```

```python
dist_of_means=[]

for i in range(1000):
    pop_sample=np.random.choice(population,10)
    dist_of_means.append(np.mean(pop_sample))
plt.hist(dist_of_means)
dist_of_means_100=[]

for i in range(1000):
    pop_sample=np.random.choice(population,100)
    dist_of_means_100.append(np.mean(pop_sample))
plt.hist(dist_of_means_100)
plt.show()
```

```python
print(np.mean(dist_of_means))
print(np.mean(dist_of_means_100))
```
