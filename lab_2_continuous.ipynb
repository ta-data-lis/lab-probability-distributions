1. Call your function with 2 sets of params below:

```python
# 1.https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.uniform.html -> rvs(loc=0, scale=1, size=1)	random variates

from scipy.stats import uniform

def rand_numb(bottom, ceiling, count):
    random_numbers = uniform.rvs(loc = bottom, scale = ceiling, size = count)
    return random_numbers

rand_numb(5, 5, 100)
```

```python
# 2.
random_100 = rand_numb(10, 15, 100)
random_1000 = rand_numb(10, 60, 1000)
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
# 1.
import numpy as np
# np.random.normal, as in class

def normal_numb(mean, std, count):
    return np.random.normal(loc = mean, scale = std, size=count)

normal_numb(10, 1, 200)
```

```python
# 2.
normal_1000 = normal_numb(10, 1, 1000)

# 3.
normal_1000_50std = normal_numb(10, 50, 1000)
```

```python
# 4.
plt.hist(normal_1000, bins = 50)
plt.xlabel("value")
plt.ylabel("count")
plt.ylim(0, 75)

plt.show()
```

```python
plt.hist(normal_1000_50std, bins = 50)
plt.xlabel("value")
plt.ylabel("count")
plt.ylim(0, 65)

plt.show()
```

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e.  = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank?

Write a code in python to solve this problem

```python
from scipy.stats import expon

lambda_inv = 10/1

expon_dist = expon(scale = lambda_inv)
print("Probability that a customer will spend less than fifteen minutes in the bank is", expon_dist.cdf(14))
```

What is the probability that the customer will spend more than 15 minutes

```python
expon_dist = expon(scale = lambda_inv)
print("Probability that a customer will spend more than fifteen minutes in the bank is", (1 - expon_dist.cdf(15)))
```
